# Math Plots

# 3-D Plots

```python
from mpl_toolkits import mplot3d
# #%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

##  $sin \sqrt{x^{2} + y^{2}}$

```python
fig = plt.figure(figsize=(12, 10))
ax = plt.axes(projection='3d')

def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax.view_init(elev=30, azim=30)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```

## $x^{2} + y^{2}$

```python
fig = plt.figure(figsize=(12, 10))
ax = plt.axes(projection='3d')

def f(x, y):
    return x ** 2 + y ** 2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax.view_init(elev=30, azim=30)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```

## Three Surface Plots

```python
fig = plt.figure(figsize=(12, 10))
ax = plt.axes(projection='3d')

def f(x, y):
    return -(x ** 2 + y ** 2)

def f2(x, y):
    return (x ** 2 + y ** 2)

def f3(x, y):
    return np.array([[1]])

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
Z2 = f2(X, Y)
Z3 = f3(X, Y)

ax.view_init(elev=30, azim=30)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.plot_surface(X, Y, Z2, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.plot_surface(X, Y, Z3, rstride=1, cstride=1,
                cmap='spring', edgecolor='none')

ax.set_title('surface');
```

# Planes

## X Varies with Y

```python
ys = np.linspace(-10, 10, 100)
zs = np.linspace(-10, 10, 100)

Y, Z = np.meshgrid(ys, zs)
X = Y
X2 = -Y
X3 = Z
X4 = -Z

fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')
#ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z)

#ax2 = fig.add_subplot(projection='3d')
ax.plot_surface(X2, Y, Z)


plt.show()
```

## X Varies with Z

```python
ys = np.linspace(-10, 10, 100)
zs = np.linspace(-10, 10, 100)

Y, Z = np.meshgrid(ys, zs)
X = Z
X2 = -Z

fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')
#ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z)

#ax2 = fig.add_subplot(projection='3d')
ax.plot_surface(X2, Y, Z)


plt.show()
```

## Y Varies

```python
xs = np.linspace(-10, 10, 100)
zs = np.linspace(-10, 10, 100)

X, Z = np.meshgrid(xs, zs)
Y = 5 - X
Y2 = 5 + X

fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')
#ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z)

#ax2 = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y2, Z)

plt.show()
```

## Z Varies

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z = X
Z2 = -X


fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')
# ax = fig.add_subplot(111, projection='3d')
ax.view_init(elev=30, azim=30)
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
plt.show()
```

## Z Varies with X and Y

```python
xs = np.linspace(0, 10, 100)
ys = np.linspace(0, 10, 100)

X, Y = np.meshgrid(xs, zs)
Z = 5 - X - .2*Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
```

## Miscellaneous Surface

```python
xs = np.linspace(0, 10, 100)
ys = np.linspace(0, 10, 100)

X, Y = np.meshgrid(xs, ys)
Z = X**2 + Y


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
```

## Sinusoidal Surface

```python
xs = np.linspace(-2.5*np.pi, 2.5*np.pi, 100)
ys = np.linspace(0, 5, 100)

X, Y = np.meshgrid(xs, ys)
Z = np.sin(X)


fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
```

## Paraboloid with Non-Uniform Scale

```python
xs = np.linspace(-4*np.pi, 4*np.pi, 100)
ys = np.linspace(0, 10, 100)

X, Y = np.meshgrid(xs, ys)
Z = X**2 + Y**2


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
```

# 2-D Plots

```python
x = np.arange(-4*np.pi,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
plt.figure(figsize=(12,6))
plt.plot(x,y,x,z)
plt.show()
```

## Decaying Cosine

```python
x = np.linspace(-2,2,41)
y = np.exp(-x**2) * np.cos(2*np.pi*x)
plt.plot(x,y,alpha=0.4,label='Decaying Cosine',
         color='red',linestyle='dashed',linewidth=2,
         marker='o',markersize=5,markerfacecolor='blue',
         markeredgecolor='blue')
plt.ylim([-2,2])
plt.legend()
plt.show()
```

## Circle

## Spiral

```python
t = np.linspace(0,6*np.pi,100)
x = t * np.cos(t)
y = t * np.sin(t)
plt.figure(figsize=(4,4))
plt.plot(x,y,alpha=1.0,linewidth=2)
#plt.ylim([-2,2])
#plt.legend()
plt.show()
```

# 3-D Parametric Plots

## 3-D Spiral

```python
plt.rcParams['legend.fontsize'] = 10
fig = plt.figure(figsize=(6, 6))

ax = fig.gca(projection='3d')

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```

## Torus

```python
# Generate torus mesh
angle = np.linspace(0, 2 * np.pi, 32)
theta, phi = np.meshgrid(angle, angle)
r, R = .25, 1.
X = (R + r * np.cos(phi)) * np.cos(theta)
Y = (R + r * np.cos(phi)) * np.sin(theta)
Z = r * np.sin(phi)

# Display the mesh
fig = plt.figure(figsize=(12, 8))
ax = fig.gca(projection = '3d')
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)
ax.plot_surface(X, Y, Z, cmap='viridis', rstride = 1, cstride = 1)
plt.show()
```
